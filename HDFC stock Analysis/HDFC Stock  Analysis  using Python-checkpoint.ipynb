```python
sns.set_style('whitegrid')
plt.title("HDFCBANK Adjusted Close Price")
HDFCBANK_df['Adj Close'].plot(figsize = (8,5))
plt.xlabel("Date")
plt.ylabel("Adjusted Close Price")
plt.grid(True)
plt.xticks(rotation = 45)
HDFCBANK_df['SMA_50'] = HDFCBANK_df['Adj Close'].rolling(window=50).mean()
HDFCBANK_df['Adj Close'].plot(figsize=(8, 5), label='Adjusted Close')
HDFCBANK_df['SMA_50'].plot(label='50-Day SMA')
plt.legend()
plt.show()
```

```python
HDFCBANK_df['Day_perc_Change'] = HDFCBANK_df['Adj Close'].pct_change()*100
```

```python
HDFCBANK_df['Day_perc_Change'] = HDFCBANK_df['Day_perc_Change'].fillna(0)
HDFCBANK_df.head()
```

```python
HDFCBANK_df['Day_perc_Change'].plot(figsize = (8,6),  fontsize = 12)
plt.title("HDFCBANK Daily Percentage Change")
plt.xlabel("Date")
plt.ylabel("Daily Percentage Change (%)", fontsize = 12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(True)
HDFCBANK_df['Day_perc_Change'].plot(figsize=(8, 6), fontsize=12, marker='o', markersize=4, color='blue')

```

```python
import numpy as np

# Step 1: Check if 'Day_Perc_Change' exists
print(HDFCBANK_df.columns)

# Step 2: Create 'Day_Perc_Change' if it doesn't exist
if 'Day_Perc_Change' not in HDFCBANK_df.columns:
    # Calculate percentage change from 'Adj Close'
    HDFCBANK_df['Day_Perc_Change'] = HDFCBANK_df['Adj Close'].pct_change() * 100

# Step 3: Define the 'trend()' function to categorize the changes
def trend(x):
    if x > -0.5 and x <= 0.5:
        return 'Slight or No change'
    elif x > 0.5 and x <= 1:
        return 'Slight Positive'
    elif x > -1 and x <= -0.5:
        return 'Slight Negative'
    elif x > 1 and x <= 3:
        return 'Positive'
    elif x > -3 and x <= -1:
        return 'Negative'
    elif x > 3 and x <= 7:
        return 'Among top gainers'
    elif x > -7 and x <= -3:
        return 'Among top losers'
    elif x > 7:
        return 'Bull run'
    elif x <= -7:
        return 'Bear drop'

# Step 4: Apply the 'trend()' function to the 'Day_Perc_Change' column
HDFCBANK_df['Trend'] = HDFCBANK_df['Day_Perc_Change'].apply(lambda x: trend(x))

# Step 5: Display the DataFrame with 'Day_Perc_Change' and 'Trend' columns
print(HDFCBANK_df[['Day_Perc_Change', 'Trend']].head())
```
